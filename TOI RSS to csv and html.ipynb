```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = "https://timesofindia.indiatimes.com/rssfeedstopstories.cms"
resp = requests.get(url)
soup = BeautifulSoup(resp.content, features="xml")
```

```python
print(soup.prettify())
```

```python
items = soup.findAll('item')
```

```python
print(items)
```

```python
len(items)
```

```python
item = items[0]
```

```python
item
```

```python
item.title.text
```

```python
news_items = []
```

```python
for item in items:
    news_item = {}
    news_item['title'] = item.title.text
    news_item['description'] = item.description.text
    news_item['link'] = item.link.text
    news_item['pubDate'] = item.pubDate.text
    news_items.append(news_item)
```

```python
print(news_items)
```

```python
import pandas as pd
df = pd.DataFrame(news_items,columns=['title','description','link','pubDate'])
```

```python
words=pd.read_csv("Desktop/New Folder/positivewords.txt",header=None, names="w")
words.head()
```

```python
df.head()
```

```python
print("List of all Positive Titles:")
ind = []
count= []
for i in range(len(df)):
    pos_words=[] #This will store the positive words in each headline
    for j in range(len(words)):
        search=words.loc[j,"w"]
        des=df.loc[i,"description"]
        split_word=des.split()
        if search in split_word:
            pos_words.append(search)
            ind.append(i)
    count.append(len(pos_words))
    if len(pos_words)>0:
        print("\nFor:",des)
        print("Words found:",pos_words)
        print("No. of words found:",len(pos_words))
print(count)
print(ind)
```

```python
unique_list = []
for x in ind:
    if x not in unique_list:
        unique_list.append(x)
print(unique_list)
```

```python
df1 = df.copy()
for i in range(0, len(df1)):
    if i not in unique_list: 
        df1 = df1.drop(i)
df1.reset_index(drop=True, inplace=True)
df1.head(10)
```
